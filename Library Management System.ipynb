```python
class ViewBook(ProjectDescription):
    def __init__(self):#         It is Contructor of Class
        self.details = "===============\nView Book Class\n===============\n\n==================\nMethods/Functions:\
\n==================\n1.viewTable"
        self.descriptions = "\n============\nDescription\n============\n1.ViewTable method just redirect \
info/Data from my sql server and display it"
        


#   Function – View()
#   creates a window for displaying the records in the table.
    def viewBook(): 
        
#       The default cursor returns the data in a tuple of tuples.
        cur = con.cursor()

        root = Tk()# creating View book window 
        root.title("Library")
        root.minsize(width=400,height=400)
        root.geometry("600x500")

        Canvas1 = Canvas(root) #Rectangular area that is intended for drawing pictures and some complex layout widgets
        Canvas1.config(bg="#12a4d9")
        Canvas1.pack(expand=True,fill=BOTH)#The geometry manager organizes widgets in 
                                            #blocks before placing them in the parent widget
#       Frame is just like a rectangular area that is used to arrrange the position of other widget in a friendy way
        headingFrame1 = Frame(root,bg="#FFBB00",bd=5)
        headingFrame1.place(relx=0.25,rely=0.1,relwidth=0.5,relheight=0.13)

        headingLabel = Label(headingFrame1, text="View Books", bg='black', fg='white', font = ('Courier',15))
#       Label widget implements a display box where you can place text or images. 
#       The text displayed by this widget can be updated at any time you want.
        headingLabel.place(relx=0,rely=0, relwidth=1, relheight=1)

        labelFrame = Frame(root,bg='black')
        labelFrame.place(relx=0.1,rely=0.3,relwidth=0.8,relheight=0.5)
        y = 0.30

        Label(labelFrame, text="%-10s%-40s%-30s%-20s"%('BID','Title','Author','Status'),
        bg='black',fg='white').place(relx=0.07,rely=0.1)
        
        Label(labelFrame, text = "================================================\
        ",bg='black',fg='white').place (relx=0.05,rely=0.2)
        
        getBooks = "select * from books;"#It is a sql statement to ridirect data from database = Library Table = books
        try:
            cur.execute(getBooks)#it will execute the statement
            con.commit()#In python autocommit feature is not available we have to manually commit after every changes


            for i in cur:#i will iterate in tuple of tuples
                Label(labelFrame,text="%-10s%-30s%-30s%-20s"%(i[0],i[1],i[2],i[3]) ,bg='black', fg='white').place(relx=0.07,rely=y)
                y += 0.1
        except:#if Some try statement fails, this block will execute and show message Box
            messagebox.showinfo("Failed to fetch files from database!!!")

        quitBtn = Button(root,text="Quit",bg='#f7f1e3', fg='black', command=root.destroy)
        quitBtn.place(relx=0.4,rely=0.9, relwidth=0.18,relheight=0.08)

        root.mainloop()
#       mainloop() is used when your application is ready to run
#       mainloop() is an infinite loop used to run the application, 
#       wait for an event to occur and process the event as long as the window is not closed.
```
